```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def trim_trailing_zeros(arr):
    # Find indices where arr is not zero
    non_zero_indices = np.where(arr != 0)[0]
    if len(non_zero_indices) == 0:
        return np.array([])  # all zeros
    last_nonzero = non_zero_indices[-1]
    return arr[:last_nonzero+1]


withFeature = np.genfromtxt("2025-12-05 12:29:18-withDependencyFeature.csv", delimiter=',')
withoutFeature = np.genfromtxt("2025-12-05 12:38:52-withoutDependencyFeature.csv", delimiter=',')

withoutFeature = trim_trailing_zeros(withoutFeature)

data = [withFeature, withoutFeature]  # skip the first element if needed

# Labels for each boxplot
xlabels = ['Pre-registered\nDependencies', 'Dependencies Created\nDuring Execution']

# Create figure and axes
fig, ax = plt.subplots(figsize=(5,5))

# Create boxplots
ax.boxplot(data,
           widths=0.4,
           medianprops=dict(color='blue', linewidth=2)
           )

# Set x-axis labels
ax.set_xticklabels(xlabels)
ax.set_yscale('log', base=10)

# Add title and labels if you want
ax.set_ylabel('Execution Time [ms]')

plt.show()
```
